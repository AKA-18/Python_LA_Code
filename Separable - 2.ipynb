```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

* Step 1- Declare the list and classes

```python
P=[1,2,3,9,10,11,17,18,19]                     #declaring the poisitve class list
N=[5,6,7,13,14,15,21,22,23]                   #decalring the negative class list
pt = P+N                                     #concatinating the list                                  
y = np.zeros(18)                              #decalring the y axis as zero
cl = ["Positive","Positive","Positive","Positive","Positive","Positive","Positive","Positive","Positive","Negative","Negative","Negative","Negative","Negative","Negative","Negative","Negative","Negative"] #decalring the class value
```

# Part A

```python
sns.scatterplot(pt,y,hue=cl)                      #plotting the 1-D scatter plot
plt.xlabel('X axis')                              #labelling the X-Axis
plt.title('Data points in 1D space')              #plot title
plt.show()                                        #show graph
```

# Part B


* The classes are not linearly  separable  in  1D  space.

# Part C

* By Shifting the Negative class by 20 and adding positive class to it to create a new list, then scatter plotting the actual point to shifted points will make it linearly seperable in 2-D plane.

```python
new_list = [x+20 for x in N]                                          #creating a new list by shifting the negative class by 20
new_list=new_list+P                                                   #adding the positive class to new_list
sns.scatterplot(pt,new_list,hue=cl)                                   #plotting the scatter plot
plt.xlabel('X axis')                                                  #labelling x axis
plt.ylabel('Y axis')                                                  #labelling yaxis
plt.title('Data points in 2D space - Linearly Seperable')             #labelling titles
plt.show()  
```
