```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

* Step 1 - Create the Dataframe

```python
# initialize list of lists
data = [[1, -1, 'Positive'], [-1,1,'Positive'], [-1,-1,'Positive'],[1,1,'Positive'],[-3,4,'Negative'],[3,-4,'Negative'],[3,4,'Negative'],[-3,-4,'Negative']]
# Create the pandas DataFrame
df = pd.DataFrame(data, columns=['X1', 'X2','Classes'])
 
# print dataframe.
df
```

# Part-A

* From the below plot we can understand that both the classes are not linearly seperable in 2D space.

# Part -B

```python
plt.rcParams['figure.figsize'] = [10, 10]                                                     #declaring the figure size                                                                                                    
plt.scatter(df['X1'].iloc[0:4,], df['X2'].iloc[0:4,], color = 'blue',label='Positive Class')  #Ploting the Positive class scatter plot
plt.scatter(df['X1'].iloc[4:8,], df['X2'].iloc[4:8,], color = 'red',label='Negative Class')   #Plotting the negative class scatter plot
plt.title('Training Data')                                                                    #plot title
plt.xlabel('X1')                                                                              #Xlabel
plt.ylabel('X2')                                                                              #Ylabel
plt.legend()                                                                                  #plotting legends
plt.show()                                                                                    #plot show 
```

# Part -C

* By Doing the second degree polynomial mapping our data becomes linearly separable (by a 2-d plane) in 3-dimensions.
