```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import svm
```

* Step 1- Declare the list and classes

```python
P=[-2, -1, 0 ,1, 2]                          #declaring the poisitve class list
N=[-8, -7, 7, 8]                             #decalring the negative class list
pt = P+N                                     #concatinating the list                                  
y = np.zeros(9)                              #decalring the y axis as zero
cl = ["Positive","Positive","Positive","Positive","Positive","Negative","Negative","Negative","Negative",] #decalring the class value
```

# Part A 

```python
sns.scatterplot(pt,y,hue=cl)                      #plotting the 1-D scatter plot
plt.xlabel('X axis')                              #labelling the X-Axis
plt.title('Data points in 1D space')              #plot title
plt.show()                                        #plotting the graph
```

# Part B

No, the Points are not Linearly Seperable in 1D Space.

# Part C 

* Taking the absolute value of both the list is one of the transformation we are doing for the 1D data to make it linearly seperable

# Part D

```python

point1 = np.array((2))                             #converting the point to np array
point2 = np.array((7))                             #converting the point to np array
dist = np.linalg.norm(point1 + point2)/2           #calculating the distance
ycl = np.zeros(1)                                  #declaring zero matrix
newcl=['Decision Point']                           #classififcation point
print('The maximum margin for linearly seperating data in 1D space is : ',dist)   #printing the distance
```

* The Equation corresponding to maximum  margin separator of  the  transformed  data  in  1D  space is 4.5

# Part E

```python
abs_p=[abs(ele) for ele in P]                                         #calculating the absolute value for positive class
abs_n=[abs(ele) for ele in N]                                         #calculating the absolute value for negative class
abs_pt = abs_p+abs_n                                                  #concatinating the list
y = np.zeros(9)                                                       #zero matrix
cl = ["Positive","Positive","Positive","Positive","Positive","Negative","Negative","Negative","Negative",] #classification value
```

```python
c=abs_pt.copy()                             #creating the copy of existing list                                           
c.append(dist)                              #appending new values
y=np.zeros(10)                              # creating zero matrix
cl = ["Positive","Positive","Positive","Positive","Positive","Negative","Negative","Negative","Negative","Decision Boundary"]   #creatinig new classification value
sns.scatterplot(c,y,hue=cl)                 #performing scatter plot
plt.xlabel('X axis')                        #labelling x axis
plt.title('Data points in 1D space')        #labeling titles
plt.show()                                  #showing plot     
```

# Part F

* Combining the actaul points and absolute points will help to convert the line into 2d plot

```python
abs_p=[abs(ele) for ele in P]                                         #calculating the absolute value for positive class
abs_n=[abs(ele) for ele in N]                                         #calculating the absolute value for negative class
abs_pt = abs_p+abs_n                                                  #concatinating the list
y = np.zeros(9)                                                       #zero matrix
cl = ["Positive","Positive","Positive","Positive","Positive","Negative","Negative","Negative","Negative",] #classification value
```

```python
sns.scatterplot(pt,abs_pt,hue=cl)                      #plotting scatterplot with actaual point and absolute points
plt.xlabel('X axis')                                   #labelling x axis
plt.ylabel('Y axis')                                   #labelling yaxis
plt.title('Data points in 2D space')                   #labelling titles
plt.show()                                             #show graph
```

# Part G

```python
PD1=pd.DataFrame(pt,columns=['X1'])                         #creating new dataframe with actual points
PD2=pd.DataFrame(abs_pt,columns=['X2'])                     #creating new dataframe with absolute points
data = ["Positive","Positive","Positive","Positive","Positive","Negative","Negative","Negative","Negative",]    #classification values
PD3 = pd.DataFrame(data, columns=['Classes'])   ##creating new dataframe with classification values
df=pd.concat([PD1,PD2,PD3],axis=1)              #Concatinating new dataframe
df                                              #showing dataframe
```

```python
x1=df.iloc[0, 0]
y1=df.iloc[4, 1]
x2=df.iloc[0, 1]
y2=df.iloc[6, 1]
```

```python
x_m_point = (x1 + x2)/2
y_m_point = (y1 + y2)/2
print();
print("The midpoint of line is :")
print( "The midpoint's x value is: ",x_m_point)
print( "The midpoint's y value is: ",y_m_point)
print();
```

* The Equation for maximum margin separator of transformed data  in 2D  space,y=0*x+4.5

# Part H

```python
X=df.iloc[:,:2]                #X-coordinate
Y=df.iloc[:,2]                 #Y-coordinate
```

```python
Linear_Model=svm.SVC(kernel='linear')      #declaring svm
```

```python
Linear_Model.fit(X,Y)                      #performing fit
```

```python
#Printing the coefficients with respect to the SVM Model

w=Linear_Model.coef_[0]                      #calculating the coefficients
m=-w[0]/w[1]
c=-Linear_Model.intercept_[0]/w[1]           #calculating the intercept
print('The equation for the maximum margin seperator is y=C')   #print statement
print('The slope m is :',m)                                     #print statement
print('The y-intercept for the line is:',c)                     #print statement
```

```python
x = [-10, 10]  # range of x values
y = [c for xx in x]  # corresponding y values

# plot the data and the hyperplane
sns.scatterplot(pt, abs_pt,hue=df['Classes'])               #scatter plotting the actual point and absolute points
plt.scatter(-2.5,4.5,color=['black'])                       #maximum margin point
plt.text(-2.5,4.5,s='Maximum Margin is [-2.5,4.5]')         #text on plot
plt.plot(x, y,label="Decision Boundary")                    #drawing the decision boundary
plt.xlabel('X1')                                       #xlabel
plt.ylabel('X2')                                       #ylabel
plt.legend()                                                #legend
plt.show()                                                  #show graph
```
